```python
import re

import sqlalchemy as sql

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../data/relatorio.csv', sep=';')
df.head()
```

```python
df['cnpj'] = df['CNPJ do gerador'].apply(lambda cnpj: ''.join(digit for digit in cnpj if digit.isdigit()))
df[['cnpj']].head()
```

```python
df_relevante = df[['cnpj', 'Quantidade Gerada', 'Unidade', 'Ano da geração']]
```

```python
df_relevante.head()
```

```python
df_relevante.describe()
```

```python
def parse_qtde(qtde):
    if not re.match(r'(\d+\.?)*\d+,\d+', qtde):
        return None
    
    return float(qtde.replace('.', '').replace(',', '.'))

df_relevante['qtde_residuo'] = df_relevante['Quantidade Gerada'].apply(parse_qtde)
df_relevante.head()
```

```python
df_agrupado_por_ano = df_relevante.groupby(['cnpj', 'Unidade', 'Ano da geração']).sum().reset_index()
df_agrupado_por_ano.head()
```

```python
df['cnpj'].nunique()
```

```python
df_agrupado_por_ano['Unidade'].unique()
```

```python
df_massa_e_volume = df_agrupado_por_ano[df_agrupado_por_ano['Unidade'] != 'Unidade'].copy()
df_massa_e_volume.head()
```

```python
df_massa_e_volume['qtde_residuo_padronizada'] = df_massa_e_volume['qtde_residuo'] / 1000
df_massa_e_volume['unidade_padronizada'] = df_massa_e_volume['Unidade'].map({'kilogramas': 'ton', 'Litro': 'm3'})
```

```python
df_massa_e_volume.head()
```

```python
df_massa_e_volume.groupby('Ano da geração').count()
```

```python
df_massa_e_volume.groupby('unidade_padronizada').count()
```

```python
df_amostra_estratificada_unidade = df_massa_e_volume.groupby('unidade_padronizada').apply(lambda group: group.sample(frac=0.01))
df_amostra_estratificada_unidade
```

```python
sns.histplot(df_amostra_estratificada_unidade, x='qtde_residuo_padronizada', hue='unidade_padronizada', bins=20)
```

```python
_, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.boxplot(df_amostra_estratificada_unidade, x='unidade_padronizada', y='qtde_residuo_padronizada', ax=axes[0])
sns.boxplot(df_amostra_estratificada_unidade, x='unidade_padronizada', y='qtde_residuo_padronizada', showfliers=False, ax=axes[1])
```

```python
df_ton = df_massa_e_volume[df_massa_e_volume['unidade_padronizada'] == 'ton']
df_m3 = df_massa_e_volume[df_massa_e_volume['unidade_padronizada'] == 'm3']
```

```python
df_ton.head()
```

```python
def filtrar_outliers_superior(df, col):
    quant25 = df[col].quantile(0.25)
    quant75 = df[col].quantile(0.75)
    iqr = quant75 - quant25
    
    return df[df[col] > quant75 + 1.5*iqr]
```

```python
df_ton_apenas_outliers = filtrar_outliers_superior(df_ton, 'qtde_residuo_padronizada')
df_ton_apenas_outliers.head()
```

```python
df_m3_apenas_outliers = filtrar_outliers_superior(df_m3, 'qtde_residuo_padronizada')
df_m3_apenas_outliers.head()
```

```python
df_outliers = pd.concat([df_ton_apenas_outliers, df_m3_apenas_outliers])
df_outliers
```

```python
df_amostra_estratificada_unidade = df_outliers.groupby('unidade_padronizada').apply(lambda group: group.sample(frac=0.01))
df_amostra_estratificada_unidade.head()
```

```python
sns.histplot(df_amostra_estratificada_unidade, x='qtde_residuo_padronizada', hue='unidade_padronizada', bins=20)
```

```python
_, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.boxplot(df_amostra_estratificada_unidade, x='unidade_padronizada', y='qtde_residuo_padronizada', ax=axes[0])
sns.boxplot(df_amostra_estratificada_unidade, x='unidade_padronizada', y='qtde_residuo_padronizada', showfliers=False, ax=axes[1])
```
