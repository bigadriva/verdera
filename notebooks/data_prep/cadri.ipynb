```python
def parse_unidade_medida(texto: str) -> str:
    if texto is None:
        return None
    
    texto = texto.split()
    texto = ''.join(texto[1:])
    return texto.strip()

df_cadri['unidade_medida'] = df_cadri['residuo_quantidade'].apply(parse_unidade_medida)
```

```python
df_cadri['unidade_medida'] = df_cadri['unidade_medida'].map({
    'm3': 'm3/ano',
    'm3/ano': 'm3/ano',
    'kg/ano': 'kg/ano',
    'L/ano': 'L/ano',
    't/ano': 'ton/ano'
})
```

```python
df_cadri['data'] = df_cadri['cadri_data'].apply(lambda date: pd.to_datetime(date, dayfirst=True))
```

```python
df_cadri[['data', 'cadri_data']].head()
```

```python
df_cadri['ano'] = df_cadri['data'].dt.year
df_cadri['ano']
```

```python
df_cadri_filtrado = df_cadri[df_cadri['unidade_medida'].isin(['ton/ano', 'm3/ano', 'kg/ano', 'L/ano'])].copy()
df_cadri_filtrado.head()
```

```python
df_cadri_filtrado.shape
```

```python
def padronizar_quantidade(linha):
    if linha['residuo_quantidade'] is None:
        return None

    residuo_quantidade = linha['residuo_quantidade'].split()[0]
    residuo_quantidade = float(residuo_quantidade)

    if linha['unidade_medida'] in ['ton/ano', 'm3/ano']:
        return residuo_quantidade
    elif linha['unidade_medida'] in ['kg/ano', 'L/ano']:
        return residuo_quantidade / 1000.0
    
    return None

def padronizar_unidade(unidade_medida):
    if unidade_medida in ['ton/ano', 'm3/ano']:
        return unidade_medida
    elif unidade_medida == 'kg/ano':
        return 'ton/ano'
    elif unidade_medida == 'L/ano':
        return 'm3/ano'

    return None
```

```python
df_cadri_filtrado['qtde_padronizada'] = df_cadri_filtrado.apply(padronizar_quantidade, axis=1)
df_cadri_filtrado['unidade_padronizada'] = df_cadri_filtrado['unidade_medida'].apply(padronizar_unidade)
```

```python
df_cadri_filtrado[['qtde_padronizada', 'unidade_padronizada']].count()
```

```python
df_cadri_filtrado[['qtde_padronizada', 'unidade_padronizada']].head()
```

```python
df_cadri_filtrado['tipo_unidade_padronizada'] = df_cadri_filtrado['unidade_padronizada'].map({'ton/ano': 'massa', 'm3/ano': 'volume'})
df_cadri_filtrado[['unidade_padronizada', 'tipo_unidade_padronizada']].head()
```

```python
df_cadri_filtrado = df_cadri_filtrado.rename(columns={'geradora_cnpj': 'cnpj'})
df_cadri_final = df_cadri_filtrado[['cnpj', 'qtde_padronizada', 'unidade_padronizada', 'tipo_unidade_padronizada', 'ano']].copy()
df_cadri_final.head()
```

```python
df_cadri_final_agg = df_cadri_final.groupby(['cnpj', 'ano', 'tipo_unidade_padronizada', 'unidade_padronizada']).sum().reset_index()
df_cadri_final_agg.head()
```

```python
def categorizar_faixa_massa_residuo(y: float):
    limites = [
        (0, 1000),
        (1000, 5000),
        (5000,10000),
        (10000,20000),
        (20000,40000),
        (40000,80000),
        (80000,160000),
        (160000,320000),
        (320000,640000),
        (640000,float('inf')),
    ]
    
    for i, limite in enumerate(limites):
        lim_inferior, lim_superior = limite
        if lim_inferior < y < lim_superior:
            return i
    return -1
```

```python
df_cadri_final_agg['faixa_massa_residuo'] = df_cadri_final_agg['qtde_padronizada'].apply(categorizar_faixa_massa_residuo)
df_cadri_final_agg.head()
```
